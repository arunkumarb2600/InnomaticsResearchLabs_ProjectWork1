```python
# Task 1: User Login Check
# stored login credentials
username = "admin"
password = "1234"
# taking input from the user
entered_username = input("Enter username: ")
entered_password = input("Enter password: ")
# checking both username and password
if entered_username == username and entered_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Task 2: Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
pass_count = 0   # to store number of passed students
fail_count = 0   # to store number of failed students
# loop through each mark
for mark in marks:
    if mark >= 50:
        pass_count += 1   # increase pass counter
    else:
        fail_count += 1   # increase fail counter
# printing the final result
print("Pass / Fail Report")
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
# Task 3: Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []  # empty list to store cleaned data
for name in names:
    name = name.strip()     # removes extra spaces from start and end
    name = name.lower()     # converts text to lowercase
    cleaned_names.append(name)  # add cleaned name to new list
print("Cleaned Names List:")
print(cleaned_names)
```

```python
# Task 4: Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
print("Message Length Analysis:")
for message in messages:
    length = len(message)  # find number of characters
    print("Message:", message)
    print("Length:", length)
    # check if message is long
    if length > 10:
        print("This message is longer than 10 characters.")
    print()  # blank line for better readability
```

```python
# Task 5: Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0  # counter for errors
for log in logs:
    if log == "ERROR":
        error_count += 1
print("Total ERROR messages:", error_count)
```
